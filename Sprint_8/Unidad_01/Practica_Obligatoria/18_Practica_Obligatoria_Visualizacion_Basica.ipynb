1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
def separa_nombre(nombre):
    nuevo = nombre[0]
    for letra in nombre[1:]:
        if letra == letra.upper():
            nuevo += " "
        nuevo += letra
    return nuevo

ciudades_mas_pobladas = df.loc[df.population_total.nlargest(5).index, 
                                    ["city","population_total"]]
ciudades = [ separa_nombre(ciudad) for ciudad in ciudades_mas_pobladas["city"].to_list()]
colores = ["blue","gray","red","olive","brown"]
plt.figure(figsize = (10,5))
plt.bar(x= ciudades, height= ciudades_mas_pobladas["population_total"]/1000, color = colores, alpha = 0.5)
plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades")

ciudades_mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(ciudades_mas_pobladas.index):
    poblacion = ciudades_mas_pobladas.loc[ciudad,"population_total"]/1000
    plt.text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center")
plt.title("Ciudades ms pobladas de California")
plt.show()
plt.savefig("ciudades_pobladas.png")
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python
plt.hist(df["area_total_km2"], color = "blue", alpha = 0.7, bins = 50,label = "Histograma")
plt.xlabel("Superficie ($km^2$)", fontdict = dict(family= "Bradley Hand itc", size = 18))
plt.ylabel("Nmero de Ciudades", fontdict= dict(family = "Bradley Hand itc", size = 18))
plt.legend()
plt.savefig("histograma.png")
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
plt.scatter(df["area_water_km2"], df["population_total"]/1000, alpha = 0.5, color = "green")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.title("Poblacin por kilometro cuadrado de agua")
outlier_1_y = ciudades_mas_pobladas.population_total.max()
outlier_2_y = ciudades_mas_pobladas.population_total.nlargest(2)[1]
outlier_1_x = df.loc[df.population_total == outlier_1_y,"area_water_km2"].values[0]
outlier_2_x = df.loc[df.population_total == outlier_2_y,"area_water_km2"].values[0]
plt.savefig("scatter_pob.png")
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(df["elevation_m"].dropna(), vert = False, boxprops = dict(facecolor = "blue"),
               medianprops = dict(color= "black"), patch_artist= True )
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.yticks([])

plt.savefig("diagrama_de_caja.png")


```
